```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('CoverImages/samim.jpeg',0)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
_,th1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
_,th2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
_,th3=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
_,th4=cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
_,th5=cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
cv2.imshow('Original ',img)
cv2.imshow('Threshhold Binary ',th1)
cv2.imshow('Threshhold Binary Inverse',th2)
cv2.imshow('Threshhold Truncated ',th3)
cv2.imshow('Threshhold TOZERO ',th4)
cv2.imshow('Threshhold TOZERO Inverse',th5)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
def nothing(x):
    print(x)
img=cv2.imread('samim.jpeg',0)
cv2.namedWindow('frame')
cv2.createTrackbar('threshold','frame',128,255,nothing)
```

```python
while (1):
    th=cv2.getTrackbarPos('threshold','frame')
    _,th1=cv2.threshold(img,th,255,cv2.THRESH_BINARY)
    _,th2=cv2.threshold(img,th,255,cv2.THRESH_BINARY_INV)
    _,th3=cv2.threshold(img,th,255,cv2.THRESH_TRUNC)
    _,th4=cv2.threshold(img,th,255,cv2.THRESH_TOZERO)
    _,th5=cv2.threshold(img,th,255,cv2.THRESH_TOZERO_INV)
    cv2.imshow('Original ',img)  
    cv2.imshow('Threshhold Binary ',th1)
    cv2.imshow('Threshhold Binary Inverse',th2)
    cv2.imshow('Threshhold Truncated ',th3)
    cv2.imshow('Threshhold TOZERO ',th4)
    cv2.imshow('Threshhold TOZERO Inverse',th5)
    
    k=cv2.waitKey(1) & 0xFF
    if k==27 :
        break
cv2.destroyAllWindows()
```

```python
img=cv2.imread('CoverImages/samim.jpeg',0)
th=100
_,th1=cv2.threshold(img,th,255,cv2.THRESH_BINARY)
_,th2=cv2.threshold(img,th,255,cv2.THRESH_BINARY_INV)
_,th3=cv2.threshold(img,th,255,cv2.THRESH_TRUNC)
_,th4=cv2.threshold(img,th,255,cv2.THRESH_TOZERO)
_,th5=cv2.threshold(img,th,255,cv2.THRESH_TOZERO_INV)
titles=['Original','THRESH_BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']
images=[img,th1,th2,th3,th4,th5]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
